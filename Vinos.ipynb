Un vino con concentración de ácido málico mayor a 3.5 se considera demasiado maduro, cuántos vinos tienen una concentración mayor?

```python
claseb=my_data[:, [0, 2]] #seleccion de las columnas 0 y 2 que contienen la clase de vino y la concentración de ácido málico
claseb
```

```python
condicionb = claseb[:,1] > 3.5 #eleccion de los valores mayores a 3,5 para la concentración de ácido málico
mask_colb = condicionb
acidom=claseb[mask_colb]
acidom
```

```python
cantidadacidom=acidom[:,1]# seleccion de los valores asociados a la la concentración de ácido málico unicamente 
cantidadacidom
```

```python
numerovinos= len(cantidadacidom) #conteo 
numerovinos
```

```python
 plt.hist(cantidadacidom)
```
